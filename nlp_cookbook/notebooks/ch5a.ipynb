```python
import re
import pandas as pd
```

```python
data_file = "../data/DataScientist.csv"
df = pd.read_csv(data_file, encoding='utf-8')
df.head()
```

```python
# Define a function that takes a column (text) and turns it into a list
def get_list_of_items(df, column_name):
    values = df[column_name].values
    values = [item for sublist in values for item in sublist]
    list_of_items = list(set(values))
    return list_of_items
```

```python
# Define a function that extracts all emails from one column of a dataframe
def get_emails(df):
    email_regex = '[^\s:|()\']+@[a-zA-Z0-9\.]+\.[a-zA-Z]+'
    df['emails'] = df['Job Description'].apply(lambda x: re.findall(email_regex, x))
    emails = get_list_of_items(df, 'emails')
    return emails
```

```python
emails = get_emails(df)
print(f"Number of emails: {len(emails)}")
print(emails)
```

```python
# Define a function that extracts all URLs from one column of a dataframe
def get_urls(df):
    url_regex = '(http[s]?://(www\.)?[A-Za-z0-9–_\.\-]+\.[A-Za-z]+/?[A-Za-z0-9$\–_\-\/\.]*)[\.)\"]*'
    df['urls'] = df['Job Description'].apply(lambda x: [
        x[item.span()[0]:item.span()[1]]
        for item in re.finditer(url_regex, x)]
    )
    urls = get_list_of_items(df, 'urls')
    return urls
```

```python
urls = get_urls(df)
print(f"Number of URLs: {len(urls)}")
print(urls)
```

```python
def get_urls2(df):
    url_regex = r'http[s]?://(?:www\.)?[A-Za-z0-9–_.\-]+\.[A-Za-z]+/?[A-Za-z0-9$–_\-/\.]*'
    df['urls'] = df['Job Description'].apply(lambda x: re.findall(url_regex, x))
    urls = get_list_of_items(df, 'urls')
    return urls
urls2 = get_urls2(df)
print(f"Number of URLs (method 2): {len(urls2)}")
print(urls2)
```

```python
# Define a function to add a new column, which measures the Levenshtein distance with an input string
def find_levenshtein(input_string, df):
    df['distance_to_' + input_string] = df['emails'].apply(
        lambda x: Levenshtein.distance(input_string, x)
    )
    return df

# Defind a function that find the closest email to an input string
def get_closest_email_lev(df, email):
    df = find_levenshtein(email, df)
    column_name = 'distance_to_' + email
    closest_email = df.loc[df[column_name].idxmin()]['emails']
    return closest_email
```

```python
new_df= pd.DataFrame(emails, columns=['emails'])
input_string = "rohitt.macdonald@prelim.com"
email = get_closest_email_lev(new_df, input_string)
print(f"The closest email to {input_string} is {email}")
```
