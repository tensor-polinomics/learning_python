```python
# Build a custom stop words list
from nltk.probability import FreqDist

def word_tokenize_nltk(text):
    return nltk.tokenize.word_tokenize(text)

def compile_stopwords_frequency(text, cut_off=0.02):
    words = word_token_nltk(text)
    freq_dist = FreqDist(word.lower() for word in words)
    words_with_freq = [
        (word, freq_dist[word]) for word in freq_dist.keys()]
    sorted_words = sorted(words_with_freq, key=lambda tup: tup[1])
    stopwords = []
    if(type(cut_off) is int):
        stopwords = [tuple[0] for tuple in sorted_words
                      if tuple[1] > cut_off]
    elif (type(cut_off) is float):
        length_cutoff = int(cut_off*len(sorted_words))
        stopwords = [tuple[0] for tuple in sorted_words[-length_cutoff:]]
    else:
        raise TypeError("cut_off should be int or float")
    return stopwords
```

```python
text = read_text_file("../data/sherlock_holmes.txt")
stopwords = compile_stopwords_frequency(text, cut_off=0.02)
print(stopwords)
print(len(stopwords))
```
