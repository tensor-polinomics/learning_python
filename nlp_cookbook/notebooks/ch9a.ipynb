```python
from transformers import (
    pipeline, BertForQuestionAnswering, BertTokenizer
)
import torch
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
qa_model = BertForQuestionAnswering.from_pretrained(
    "bert-large-uncased-whole-word-masking-finetuned-squad"
).to(device)
qa_tokenizer = BertTokenizer.from_pretrained(
    "bert-large-uncased-whole-word-masking-finetuned-squad"
)
qa_pipeline = pipeline(
    "question-answering",
    model=qa_model,
    tokenizer=qa_tokenizer,
    device=0 if torch.cuda.is_available() else -1,
)
```

```python
context = """
Artificial intelligence is revolutionizing the financial services industry by 
enhancing decision-making, risk management, and customer experiences. Machine learning 
algorithms now power sophisticated fraud detection systems that analyze millions of 
transactions in real-time, identifying suspicious patterns with unprecedented accuracy. 
Investment firms leverage AI-driven quantitative models to predict market movements 
and optimize portfolio allocations, while robo-advisors democratize wealth management 
by providing personalized investment strategies to retail clients at minimal cost. 
Natural language processing enables automated analysis of earnings calls, 
news sentiment, and regulatory filings, giving traders valuable insights within 
milliseconds. Beyond trading and investments, AI streamlines back-office operations 
through intelligent document processing, automates credit scoring with alternative 
data sources, and powers chatbots that handle routine customer inquiries 24/7. 
However, the integration of AI in finance also raises important concerns around 
algorithmic bias, model interpretability, and systemic risk, prompting regulators 
to develop frameworks ensuring these powerful technologies are deployed responsibly 
and transparently.
"""
```

```python
question = "What is the impact of artificial intelligence in finance?"
result = qa_pipeline(question=question, context=context)
print(f"Answer: {result['answer']}")
```
